# akshare 常用方法示例

akshare 是一个开源的 Python 金融数据接口库，聚合了东方财富、新浪、同花顺等数据源。

官方文档：https://akshare.akfamily.xyz/

```python
import os
os.environ['NO_PROXY'] = '*'

import akshare as ak
import pandas as pd
```

---
## 1. 获取沪深股票列表

```python
# 上交所股票列表
stock_sh_1 = ak.stock_info_sh_name_code(symbol="主板A股")
stock_sh_2 = ak.stock_info_sh_name_code(symbol="主板B股")
stock_sh_3 = ak.stock_info_sh_name_code(symbol="科创板")
stock_sh = pd.concat([stock_sh_1, stock_sh_2, stock_sh_3], ignore_index=True)
print(f"上交所股票数量: {len(stock_sh)}")
print(stock_sh_1.head())
print(stock_sh_2.head())
print(stock_sh_3.head())
```

```python
# 深交所股票列表
stock_sz_1 = ak.stock_info_sz_name_code(symbol="A股列表")
stock_sz_2 = ak.stock_info_sz_name_code(symbol="B股列表")
stock_sz_3 = ak.stock_info_sz_name_code(symbol="AB股列表")
stock_sz_4 = ak.stock_info_sz_name_code(symbol="CDR列表")
stock_sz = pd.concat([stock_sz_1, stock_sz_2, stock_sz_3, stock_sz_4], ignore_index=True)
print(f"深交所股票数量: {len(stock_sz)}")
print(stock_sz_1.head())
print(stock_sz_2.head())
print(stock_sz_3.head())
print(stock_sz_4.head())
```

## 2. 个股日K线数据 (stock_zh_a_hist)

获取单只股票的历史日K线行情数据，数据来源：东方财富

- `symbol`: 股票代码（不带后缀）
- `period`: 周期，`daily`/`weekly`/`monthly`
- `start_date` / `end_date`: 日期范围，格式 `YYYYMMDD`
- `adjust`: 复权方式，`""`=不复权, `"qfq"`=前复权, `"hfq"`=后复权

```python
# 获取平安银行(000001)日K线，前复权
df_kline = ak.stock_zh_a_hist(
    symbol="000001",
    period="daily",
    start_date="20260701",
    end_date="20260711",
    adjust="qfq"
)
print(f"数据条数: {len(df_kline)}")
print(f"列名: {list(df_kline.columns)}")
print(df_kline.tail(10))
```

```python
# 获取贵州茅台(600519)周K线，不复权
df_weekly = ak.stock_zh_a_hist(
    symbol="600519",
    period="weekly",
    start_date="20260701",
    end_date="20260711",
    adjust=""
)
print(df_weekly.tail(5))
```

## 3. A股实时行情快照 (stock_zh_a_spot_em)

获取全部A股的实时行情数据（当日快照），包含最新价、涨跌幅、成交量、换手率、市盈率、市净率、总市值等

```python
# 获取全部A股实时行情
df_spot = ak.stock_zh_a_spot_em()
print(f"股票数量: {len(df_spot)}")
print(f"列名: {list(df_spot.columns)}")
print(df_spot.head(5))
```

```python
# 筛选涨幅前10的股票
df_top = df_spot.nlargest(10, '涨跌幅')[['代码', '名称', '最新价', '涨跌幅', '成交额', '市盈率-动态', '总市值']]
print("今日涨幅前10:")
print(df_top.to_string(index=False))
```

## 14. 个股估值数据 (stock_zh_valuation_baidu)

获取个股历史估值指标（来源：百度股市通），包含总市值、市盈率(TTM)、市净率等

- `symbol`: 股票代码（不带后缀）
- `indicator`: 指标名，如 `总市值`、`市盈率(TTM)`、`市净率`
- `period`: 时间范围，如 `近一年`、`近三年`、`近五年`

```python
# 贵州茅台估值数据（百度股市通）
for indicator in ['总市值', '市盈率(TTM)', '市净率']:
    df_val = ak.stock_zh_valuation_baidu(symbol="600519", indicator=indicator, period="近一年")
    print(f"--- {indicator} ---")
    print(f"  数据条数: {len(df_val)}, 最新值: {df_val.iloc[-1]['value']}")
    print(df_val.tail(3))
```
